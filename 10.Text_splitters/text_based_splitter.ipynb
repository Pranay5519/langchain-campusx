# TextBased SPlitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
Large Language Models (LLMs) are advanced AI systems based on neural networks and transformer architectures, trained on massive datasets to understand and generate human-like text. They power applications like chatbots (ChatGPT, Gemini), translation, and content summarization. LLMs use self-attention to identify context and relationships within data.
Key Aspects of LLMs:
Core Technology: Built on transformers, which are deep learning models that use parallel processing for efficiency.
Training Process: They are pre-trained on vast datasets and often fine-tuned for specialized tasks.
Key Capabilities: Text generation, summarization, question answering, translation, and coding assistance.
Limitations: They can suffer from hallucinations (generating false info), biases from training data, and lack true understanding of context.
Popular Examples: OpenAI's GPT-4o, Google's Gemini, Anthropic's Claude, and IBM Granite. 

"""
# Initialize the splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=130,
    chunk_overlap=0,
)

# Perform the split
chunks = splitter.split_text(text)
len(chunks)
```

```python
len(chunks[0]) , len(chunks[1]) , len(chunks[2])
```

```python
for chunk in chunks:
    print(chunk)
    print("----")
```

```python
# Splitting Code
from langchain_text_splitters import RecursiveCharacterTextSplitter 
text = """
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade  # Grade is a float (like 8.5 or 9.2)

    def get_details(self):
        return self.name"

    def is_passing(self):
        return self.grade >= 6.0


# Example usage
student1 = Student("Aarav", 20, 8.2)
print(student1.get_details())

if student1.is_passing():
    print("The student is passing.")
else:
    print("The student is not passing.")

"""

# Initialize the splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=0,
)

# Perform the split
chunks = splitter.split_text(text)
len(chunks)
```

```python
len(chunks[0]) , len(chunks[1]) , len(chunks[2])
```
